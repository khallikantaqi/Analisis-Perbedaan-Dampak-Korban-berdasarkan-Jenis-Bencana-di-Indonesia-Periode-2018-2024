### 4. Data Cleaning

```python
df = df.copy()
```

```python
df[df.duplicated()]
```

Kita perlu melakukan pengecekan terlebih dahulu, apakah terdapat data duplikat pada datset ini, karena data duplikat nanti dapat menyebabkan analisis yang tidak akurat.

```python
df.duplicated().sum()
```

Ditemukan terdapat 329 data duplikat pada dataset ini.

```python
df = df.drop_duplicates()
```

Dilakukan penghapusan data duplikat menggunakan .drop_duplicates().

```python
df.duplicated().sum()
```

Dilakukan pengecekan kembali untuk data duplikat, dan berdasarkan penaganan yang telah dilakukan data duplikatnya sekarang menjadi "0".

```python
df.isnull().sum()
```

Dilakukan pengecekan missing values dan masih terdapat missing values pada kolom "death","missing_person","flooded_house".

```python
num = ["death","missing_person","flooded_house"] 

df[num] = df[num].fillna(0)
# df.loc[:, num] = df.loc[:, num].fillna(0)
```

Missing values diisi dengan nilai 0 dengan asumsi tidak terdapat korban atau dampak pada kejadian tersebut.

```python
df.isnull().sum()
```

Dilakukan pengecekan ulang dan pengisian missing values telah berhasil dilakukan karena missing values pada semua kolom sekarang menjadi "0".

```python
# ubah ke datetime
df["date"] = pd.to_datetime(df["date"])

# ubah ke integer
cols = ["death","missing_person","injured_person","damaged_house","flooded_house"]
df[cols] = df[cols].astype(int)
```

- Tipe data "date" dirubah menjadi datetime
- Beberapa kolom numerik yang tadinya bertipe data float dirubah ke tipe data integer

```python
df.info()
```

Dilakukan pengecakan ulang, dan penggantian tipe data telah kolom "date" menjadi date time dan kolom numerik menjadi integer telah berhasil dilakukan.

```python
df["disaster_type"].value_counts()
```

Dilakukan pengecekan kembali setelah dilakukan handling data yang terduplikat.

```python
df["disaster_type"] = df["disaster_type"].replace({
    "GELOMBANG PASANG DAN ABRASI": "GELOMBANG PASANG / ABRASI",
    "LETUSAN GUNUNG API": "ERUPSI GUNUNG API"
})
```

Dari inkonsistensi akan dilakukan standarisasi jenis bencana dengan menggabungkan jenis bencana yang serupa yaitu menjadi 'GELOMBANG PASANG / ABRASI' dan 'ERUPSI GUNUNG API' 

```python
df["disaster_type"].value_counts()
```

```python
df["year"] = df["date"].dt.year
```

Kolom "year" akan ditambahkan dari ekstraksi kolom "date". Hal ini dilakukan untuk memudahkan melihat periode tahun (2018-2024) tanpa melihat tanggal atau bulannya, sehingga analisis nantinya sudah berfokus pada periodenya tahunnya saja.

##### Uji Statistik Inferential

Menguji Hipotesis

Untuk mengetahui apakah terdapat perbedaan rata-rata jumlah korban berdasarkan jenis bencana di Indonesia digunakan metode ANOVA.

Hipotesis yang digunakan untuk setiap variabel korban:
- H0 (Hipotesis nol) : Tidak terdapat perbedaan rata-rata jumlah korban antar jenis bencana.
- H1 (Hipotesis alternatif) : Terdapat perbedaan rata-rata jumlah korban antar jenis bencana.

Uji ANOVA dipilih karena:
- Tujuan analisis adalah membandingkan rata-rata jumlah korban antar beberapa kelompok (jenis bencana), 
- Tipe data yang sesuai kategorikal (disaster_type) dengan numerik (death, injured_person, missing_person),
- Jenis bencana terdiri dari banyak kategori
- Dapat menghandle jumlah data besar

```python
# death
groups = [group["death"].values for name, group in df.groupby("disaster_type")]
stat, p_val = f_oneway(*groups)
print(f"Korban Meninggal p-value: {p_val}")

# injured_person
groups = [group["injured_person"].values for name, group in df.groupby("disaster_type")]
stat, p_val = f_oneway(*groups)
print(f"Korban Luka p-value: {p_val}")

# missing_person
groups = [group["missing_person"].values for name, group in df.groupby("disaster_type")]
stat, p_val = f_oneway(*groups)
print(f"Korban Hilang p-value: {p_val}")
```

Berdasarkan hasil uji ANOVA diperoleh p-value dari seluruh variabel = 0.0

P-value sleuruh variabel = 0.0 ini bukan tidak ada/bernilai 0 namun sangat kecil sekali 

Karena p-value < 0.05, maka:
- H0 ditolak
- Terdapat perbedaan signifikan rata-rata jumlah korban antar jenis bencana
- Hal ini berlaku untuk semua variabel (Korban meninggal, luka, dan hilang)

Hasil uji inferensial menunjukkan bahwa jenis bencana memiliki pengaruh signifikan terhadap seluruh jenis dampak korban (korban meninggal, luka, dan hilang). Dari hasil juga menegaskan bahwa perbedaan dampak korban antar jenis bencana dipengaruhi oleh karakteristik masing-masing bencana.
